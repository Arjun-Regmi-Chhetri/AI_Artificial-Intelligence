```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as split
from sklearn.linear_model import LinearRegression as lr
```

```python
dataset = pd.read_csv("../salary_data.csv")
dataset = dataset.dropna()
dataset
```

```python
dataset = dataset[['Years of Experience', 'Salary']]
dataset
```

```python
x = dataset.iloc[:, : -1].values
x
```

```python
y = dataset.iloc[:, 1:].values
y
```

```python
x_train, x_test, y_train, y_test = split(x,y, test_size= 0.1, random_state=1)
```

```python
# checking length of all 

print(" Total length : " , len(x))
print(" Total length of x_train : ", len(x_train))
print( " Total length of y_train : ", len(x_test))
```

```python
# training model 
model = lr()
model.fit(x_train, y_train)
```

```python
# testing model

def predict():
    exp = float(input("Enter a Year of Experience : "))
    exp = np.array([[exp]])

    print(exp)


    #predict salary

    predict_salary = model.predict(exp)
    print(predict_salary)

predict()

```
